```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import mannwhitneyu
# посчитанные МДА (минимальная детектируемая активность)
sample1 = np.array([137.0170593, 188.4097675, 207.0290316, 151.2095629, 36.92627781, 243.85830147, 329.947148001, 151.8410329477, 271.927613024])
sample2 = np.array([60.82342534, 108.6498529, 220.26976542, 108.00722672, 200.844623653, 121.3291858302, 81.3904810453, 94.6801942957, 12.940243762])
plt.figure(figsize=(8, 5))
plt.scatter(['КГС'] * len(sample1), sample1, label='Ксеноновый детектор', s=60)
plt.scatter(['Сцинт.'] * len(sample2), sample2, label='Сцинтиллятор', s=60)
plt.ylabel('Значения МДА')
plt.title('Сравнение двух выборок')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

statistic, p_value = mannwhitneyu(sample1, sample2, method='exact')
print(f"\nРезультат теста Манна-Уитни: p = {p_value:.4f}")
```
